```python
import matplotlib.pyplot as plt
```

```python
categories = ['A','B','C']
values_1=[10,15,12]
values_2=[8,17,14]
```

```python
plt.plot(categories, values_1, marker='o',label='2023')
plt.plot(categories, values_2, marker='o',label='2024')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title('Slopegraph')
plt.legend()
plt.show()
```
