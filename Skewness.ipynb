```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.Series([1,2,3,4,5,6,7,8,9])
```

```python
skewness = data.skew()
print('skewness:',skewness)
```

```python
sns.distplot(data)
plt.title("data distribution")
plt.xlabel("value")
plt.ylabel("frequency")
plt.show()
    
```
