```python
import matplotlib.pyplot as pyplot
```

```python
labels =("python","scala","c#","java","php")
index = (1,2,3,4,5)
sizes=[45,10,15,30,22]
```

```python
pyplot.bar(index,sizes, tick_label=labels, color=("red","green","blue","yellow","orange"))

pyplot.ylabel("usage")
pyplot.xlabel("program languages")

pyplot.show()
```
