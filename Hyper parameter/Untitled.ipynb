```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd

from matplotlib  import pyplot as plt

import seaborn

from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier

from sklearn.naive_bayes import MultinomialNB

from sklearn.tree import DecisionTreeClassifier

from sklearn.linear_model import LogisticRegression

from sklearn.svm import SVC

from sklearn.datasets import load_digits

from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
digits=load_digits()
print(digits)
```

```python
data= pd.DataFrame(digits.data)
data.head()
```

```python
data.shape
```

```python
data['target']=digits.target
data.head()
```

```python
x=data.drop(['target'],axis=1)
y=data.target
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2)
```

```python
print(xtrain.shape)
```

```python
print(ytrain.shape)
```

```python
print(xtest.shape)
```

```python
print(ytest.shape)
```

```python
model_params = {
    'SVM': {
        'model': SVC(gamma='auto'),
        'params': {
            'C': [0.1, 1, 5, 10],
            'kernel': ['rbf', 'linear']
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(),
        'params': {
            'C': [0.1, 1, 5, 10]
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [10, 50, 100],
            'max_depth': [None, 10, 20, 30]
        }
    },
    'naive_bayes': {
        'model': MultinomialNB(),
        'params': {
            'alpha': [0.1, 1, 5, 10]
        }
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'max_depth': [None, 10, 20, 30],
            'min_samples_split': [2, 5, 10]
        }
    }
}
```

```python
scores=[]
for model_name in model_params:
    mp=model_params[model_name]
    gds=GridSearchCV(mp['model'],mp['params'],cv=5)
    gds.fit(digits.data,digits.target)
    scores.append({'model':model_name,'best_score':gds.best_score_,
                'best_params':gds.best_params_})
```

```python
scores
```

```python
df=pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

##### The best model is SVM with parameters C = 0.1 and Linear kernel.

```python
model_final = SVC(C=0.1,kernel='linear')
```

```python
model_final.fit(xtrain,ytrain)
```

```python
ytrain_pred = model_final.predict(xtrain)
ytrain_pred
```

```python
train_accuracy = accuracy_score(ytrain,ytrain_pred)
train_accuracy
```

```python
ytest_pred=model_final.predict(xtest)
ytest_pred
```

```python
test_accuracy=accuracy_score(ytest,ytest_pred)
test_accuracy
```
