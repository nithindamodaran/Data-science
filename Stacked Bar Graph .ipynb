```python
import matplotlib.pyplot as pyplot
```

```python
labels=('python','scala','c#','java','php')
index=(1,2,3,4,5)
web_usage=[20,2,5,10,14]
data_science_usage=[15,8,5,15,2]
game_usage=[10,1,5,5,4]
```

```python
pyplot.bar(index,web_usage,tick_label=labels,label='web')
pyplot.bar(index, data_science_usage, tick_label=labels,label='data science',bottom=web_usage)
web_and_game_usage=[web_usage[i]+data_science_usage[i]
for i in range(0, len(web_usage))]
pyplot.bar(index, game_usage, tick_label=labels, label='game',bottom=web_and_game_usage)

pyplot.ylabel('usage')
pyplot.xlabel('programing languages')
pyplot.legend()
pyplot.show()
```
